```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas  as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
plt.rcParams['figure.figsize'] = [30,15]
```

```python
df = pd.read_csv('Salary_Data.csv')
df.head(20)
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, 1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
pred = pd.DataFrame(y_pred)
pred
```

```python
# Plotted the actual & predicted 
plt.scatter(x_train, y_train, color = 'red', s = 200)
plt.plot(x_train, lr.predict(x_train), color = 'blue')
plt.xticks(fontsize =25,rotation = 0)
plt.yticks(fontsize = 25)
plt.title('Salary vs Experience (Actual vs. Predicted)', fontsize = 50)
plt.xlabel('Years of Experience',fontsize = 30)
plt.ylabel('Salary',fontsize = 30)
plt.show()
```

```python
# Output shows that the predicted salary when the years of experience exceeds 10
i = 10
while i <= 20:
    print("The salary for %d ==>" %(i) , int(lr.predict([[i]])))
    i= i+1
```
