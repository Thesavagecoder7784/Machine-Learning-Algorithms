```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.datasets import load_breast_cancer
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
```

```python
# Classification Task: Breast Cancer Dataset
# Load dataset
data_class = load_breast_cancer()
X_class = data_class.data
y_class = data_class.target
```

```python
data_class['data']
```

```python
# Split the dataset into train and test sets
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(
    X_class, y_class, test_size=0.2, random_state=42
)

# Initialize the XGBoost classifier
xgb_class = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    use_label_encoder=False,
    eval_metric="logloss",
    random_state=42
)

# Fit the model
xgb_class.fit(X_train_class, y_train_class)
```

```python
# Make predictions
y_pred_class = xgb_class.predict(X_test_class)

# Evaluate the model
accuracy = accuracy_score(y_test_class, y_pred_class)
print(f"Classification Accuracy: {accuracy:.2f}")
```

```python
# Perform 5-fold cross-validation
cv_scores = cross_val_score(xgb_class, X_class, y_class, cv=5, scoring='accuracy')
print(f"Cross-Validation Accuracy: {cv_scores.mean():.2f} ± {cv_scores.std():.2f}")
```
